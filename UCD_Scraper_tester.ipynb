```python
#Only good for static websites
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://catalog.ucdavis.edu/departments-programs-degrees/computer-science-engineering/computer-science-bs/#requirementstext"
base_url = "https://catalog.ucdavis.edu"

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
#<table class = "sc_courselist">
#soup.find('table', class_ = 'sc_courselist')
#Course = codecol
#Class name = titlecol
#Units (in href) = 
#links = href
#Plan: Go through every link, and extract the paragraph for every class

table = soup.find('table')
```

```python
#This code gets you the Course Name, Title, the Units, Course Description, Prerequisites, Learning Activites, Credit Limitations, Grade Mode, and General Education

world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text for title in world_titles]

print(world_table_titles)
```

```python
df = pd.DataFrame(columns= world_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data:
    row_data = row.find_all('td')
    print(row_data)
    individual_row_data = [data.text.strip().replace('\xa0', ' ') for data in row_data]
    if len(individual_row_data) == 3:
        length = len(df)
        df.loc[length] = individual_row_data

    # print(individual_row_data)




```
